```python
# Import required libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns
from datetime import datetime

from scipy.stats import beta
from scipy.optimize import minimize
from tqdm import tqdm
from math import log

# Suppress warnings to ensure a clean output
import warnings
warnings.filterwarnings('ignore')

# Load the dataset for Subscriptions 
# The dataset is sorted in the following hierarchical order:
# 1. Cohort - Ascending order.
# 2. Customer_ID - Ascending order within each cohort.
# 3. Renewal_Period_No - Ascending order within each customer.
# This sorting ensures that the data is organized first by cohort groups,
# then by individual customers within each cohort, and finally by the renewal
# period number for each customer.
df = pd.read_csv(r"data\Subscribers.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Identify unique cohorts from the 'Cohort' column in the dataframe
selected_cohorts = df['Cohort'].unique().tolist()
```

```python
def calculate_rates(df_prep, selected_cohort):
    """
    Calculates the retention and survival rates for a given cohort.
    
    Parameters:
        df_prep (pd.DataFrame): The preprocessed DataFrame - Subscriptions data.
        selected_cohort (str): The cohort for which retention and survival rates are calculated.
    
    Returns:
        tuple: A tuple containing the cohort, number of customers (sum_customers), retention rates, and survivor rates.
    """
    # Filter the DataFrame for the selected cohort and reset the index
    filtered_df = df_prep[df_prep['Cohort'] == selected_cohort].reset_index(drop=True)
    # Select relevant columns and reset index again for clean processing
    relevant_data = filtered_df[['Customer_ID', 'Renewal_Period_No', 'Subscription_Status']].reset_index(drop=True)
    # Pivot the DataFrame to have Customer_ID as index, Renewal_Period_No as columns
    pivot_df = relevant_data.pivot(index='Customer_ID', columns='Renewal_Period_No', values='Subscription_Status')
    pivot_df.reset_index(inplace=True)
    # Remove the axis name for clarity
    pivot_df.columns.name = None
    # Ensure column names are string for consistency
    pivot_df.columns = pivot_df.columns.astype(str) 
    # Replace NaN values with 0
    pivot_df.replace(np.nan, 0, inplace=True)  
    # Convert columns (except 'Customer_ID') to integer for processing
    pivot_df.loc[:, pivot_df.columns != 'Customer_ID'] = pivot_df.loc[:, pivot_df.columns != 'Customer_ID'].astype(int)
    
    # Summarize data to get the total counts per renewal period
    # Exclude the 'Customer_ID' column
    sum_column = pivot_df.sum(axis=0)[1:]  
    sum_customers = sum_column.tolist()
    
    # Calculating retention rate
    def ShiftLeft(L0):
        L=L0.copy()
        g=L[0]
        L.pop(0)
        L.append(g)
        return(L)
    
    retention_rates= [i / j for i, j in zip(ShiftLeft(sum_customers), sum_customers)]
    retention_rates.pop(-1);
    
    # Calculating survivor function  
    survivor_rates= [i / j for i, j in zip(sum_customers, [sum_customers[0]]*len(sum_customers))]
    
    return (selected_cohort, sum_customers, retention_rates, survivor_rates)
```

```python
# Analyze retention rates and survivor function for each cohort
sum_customers_many = []
retention_rates_many = []
survivor_rates_many = []

for cohort in selected_cohorts:
    retention_data = calculate_rates(df, cohort)
    sum_customers_many.append(retention_data[1])
    retention_rates_many.append(retention_data[2])
    survivor_rates_many.append(retention_data[3])
```

```python
# Cohort Analysis
# Convert the list of retention rates into a 2D numpy array, handling varying lengths
retention_matrix = np.array([np.pad(array, (0, max(len(a) for a in retention_rates_many) - len(array)), 'constant', constant_values=np.nan) for array in retention_rates_many])

# Generate cohort labels and renewal labels for the heatmap
cohorts_labels = selected_cohorts
renewals_labels = [f'Renewal {i+1}' for i in range(retention_matrix.shape[1])]

# Convert the numpy array to a pandas DataFrame for plotting
retention_df = pd.DataFrame(retention_matrix, index=cohorts_labels, columns=renewals_labels)

# Create the heatmap of retention rates
plt.figure(figsize=(10, 8))
sns.heatmap(retention_df, annot=True, fmt=".2f", cmap="gray")
plt.title('Cohort Retention Rates Heatmap')
plt.xlabel('Renewal Period')
plt.ylabel('Cohort')
plt.show()
```

```python
# Beyond Observed Survival - Model the future
def MC_sBG_func(surv_many, ret_many, F_horizon, period_freq, PDF_status):
    """
    Model future survival and retention rates using a stochastic Beta-Geometric (sBG) model.
    
    Parameters:
    - surv_many: List of lists containing survival rates for multiple cohorts.
    - ret_many: List of lists containing retention rates for multiple cohorts.
    - F_horizon: Integer representing the number of periods to forecast into the future.
    - period_freq: String indicating the time unit for forecasting ('MS' for months, etc.).
    - PDF_status: String, if 'ON', generates PDF plots of the Beta distribution and future projections.
    
    Returns:
    Tuple containing time periods, predicted survival rates, predicted retention rates, 
    Alpha and Beta parameters, and expected average tenure of a customer.
    """
    
    def generate_probabilities(alpha, beta, x):
        """Generate probabilities for a given number of periods using alpha and beta parameters."""
        p = [alpha / (alpha + beta)]
        for t in range(1, x):
            pt = (beta + t - 1) / (alpha + beta + t) * p[t-1]
            p.append(pt)
        return p
    
    def survivor(probabilities, t):
        """Calculate the survivor function S at time t."""
        s = 1 - probabilities[0]
        for x in range(1, t + 1):
            s -= probabilities[x]
        return s
    
    def log_likelihood_multi_cohort(alpha, beta, data):
        """Calculate the log-likelihood across multiple cohorts for given alpha and beta."""
        if alpha <= 0 or beta <= 0:
            return -1000
        probabilities = generate_probabilities(alpha, beta, len(data[0]))
        total = 0
        for i, cohort in enumerate(data):
            total += sum([(cohort[j]-cohort[j+1])*log(probabilities[j]) for j in range(len(cohort)-1)])
            total += cohort[-1] * log(survivor(probabilities, len(data) - i - 1))
        return total
    
    def maximize_multi_cohort(data):
        """Maximize the log-likelihood function to find the optimal alpha and beta."""
        func = lambda x: -log_likelihood_multi_cohort(x[0], x[1], data)
        x0 = np.array([1., 1.])
        res = minimize(func, x0, bounds=[(0, None), (0, None)], options={'ftol': 1e-8})
        return res
    
    def predicted_retention(alpha, beta, t):
        """Calculate the predicted retention rate at time t based on alpha and beta."""
        return (beta + t) / (alpha + beta + t)

    def predicted_survival(alpha, beta, x):
        """Calculate the predicted survival rates over a range of periods."""
        s = [predicted_retention(alpha, beta, 0)]
        for t in range(1, x):
            s.append(predicted_retention(alpha, beta, t) * s[t-1])
        return s
    
    def fit_multi_cohort(data):
        """Fit the model."""
        res = maximize_multi_cohort(data)
        if res.status != 0:
            raise Exception(res.message)
        return res.x

    # Fit model and calculate predictions/projections
    a, b = fit_multi_cohort(surv_many)
    t_tot = range(1, len(surv_many[0][1:]) + 1 + F_horizon)
    surv_new = predicted_survival(a, b, len(surv_many) + F_horizon)
    ret_new = [predicted_retention(a, b, i) for i in t_tot]

    # Plotting and area calculation
    if PDF_status == 'ON':
        plot_beta_distribution(a, b)
        plot_future_projections(surv_new, surv_many)

    # expected (average) tenure of a customer
    ETC = calculate_expected_tenure(t_tot, surv_new, period_freq)
    
    return t_tot, surv_new, ret_new, a, b, ETC
```

```python
def plot_beta_distribution(a, b):
    """Plot the Beta distribution given alpha and beta parameters."""
    x = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
    plt.figure(figsize=(7, 4))
    plt.plot(x, beta.pdf(x, a, b), 'r-')
    plt.title('Beta Distribution', fontsize=15)
    plt.xlabel('Values of Churn Probabilities', fontsize=15)
    plt.ylabel('Probability Density', fontsize=15)
    plt.xlim(0,1)
    plt.show()

def plot_future_projections(surv_new, surv_many):
    """Plot future survival rate projections."""
    plt.figure(figsize=(7, 4))
    plt.plot(surv_new)
    plt.ylabel('Percent Surviving', fontsize=15)
    plt.xlabel('Renewal Periods (Semi-years)', fontsize=15)
    plt.axvline(x=len(surv_many[0][1:]), color='r', linewidth=1, linestyle='dashed')
    plt.show()

def calculate_expected_tenure(t_tot, surv_new, period_freq):
    """Calculate the expected average tenure of a customer."""
    return integrate(t_tot, surv_new) * (int(period_freq.replace("MS", "")) / 12)

def integrate(x, y):
    """Perform numerical integration of y with respect to x to find the area under the curve (sm)."""
    sm = 0
    for i in range(1, len(x)):
        h = x[i] - x[i-1]
        sm += h * (y[i-1] + y[i]) / 2
    return sm
```

```python
period_freq='6MS' # Renewal periods
F_horizon=6 # Number of renewal cycles projected into the future
mc_sBG_output=MC_sBG_func(survivor_rates_many,retention_rates_many,F_horizon,period_freq,'ON')
print('The expected (average) tenure of a customer in this cohort is: ',round(mc_sBG_output[5],2), 'Years')
print(f'Alpha: {round(mc_sBG_output[3],2)}, Beta: {round(mc_sBG_output[4],2)}')
```

```python
def calculate_hv(survivor_rates, discount_rate, net_cashflow, selected_cohort, data):
    # Filter customers based on the selected cohort
    cohort_customers = data[data['Cohort'] == selected_cohort]
    
    # Create discount and net cashflow lists
    discounts = [discount_rate] * len(survivor_rates)
    net_cashflows = [net_cashflow] * len(survivor_rates)
    time_periods = np.arange(len(survivor_rates))
    
    # Calculate discount factors
    discount_factors = [(1 / (1 + discount_rate) ** t) for t in time_periods]
    
    # Get unique customer IDs for the selected cohort
    unique_customers = cohort_customers['Customer_ID'].unique().tolist()
    
    # Initialize DataFrame for Historical Value (HV) calculations
    df_hv = pd.DataFrame()
    hv_values = []
    customer_ids = []
    
    # Calculate Historical Value for each customer
    for customer_id in tqdm(unique_customers):
        customer_ids.append(customer_id)
        subscription_statuses = cohort_customers[cohort_customers['Customer_ID'] == customer_id]['Subscription_Status'].tolist()
        
        hv = sum(
            net_cashflow * discount_factor * subscription_status
            for net_cashflow, discount_factor, subscription_status in zip(net_cashflows, discount_factors, subscription_statuses)
        )
        hv_values.append(hv)
        
    # Populate the DataFrame with Historical Value results
    df_hv['Customer_ID'] = customer_ids
    df_hv['HV'] = hv_values
    df_hv['HV'] = df_hv['HV'].round(1)
    df_hv['Cohort'] = selected_cohort
    
    return df_hv
```

```python
# Parameters
discount_rate = 0.05
net_cashflow = 10  # Expected value (or net cashflow) of customer - subscription rate

# Calculate Historical Value (HV)
hv_output = pd.DataFrame()
for cohort, survivor_rate in zip(selected_cohorts, survivor_rates_many):
    hv_output_temp = calculate_hv(survivor_rate, discount_rate, net_cashflow, cohort, df)
    hv_output = pd.concat([hv_output, hv_output_temp]).reset_index(drop=True)

print('The Historical Value table is:')
print(hv_output)
```

```python
def calculate_rclv(survivor_rates_model, discount_rate, net_cashflow, selected_cohort, data, forecast_horizon):

    # Convert to datetime object
    date_obj = datetime.strptime(selected_cohorts[-1], '%Y-%m')

    # Format the date in the desired format and manually remove leading zeros
    last_cohort_date = date_obj.strftime('%m/%d/%Y').replace('/0', '/').lstrip('0')
    
    # Get active customers in the last cohort period
    active_customers_list = data[(data['Period_Start'] == last_cohort_date) & (data['Subscription_Status'] == 1)]['Customer_ID'].unique().tolist()
    active_customers = data[data['Customer_ID'].isin(active_customers_list)]
    
    # Create discount and net cashflow lists
    discounts = [discount_rate] * forecast_horizon
    net_cashflows = [net_cashflow] * forecast_horizon
    time_periods = np.arange(forecast_horizon).tolist()
    
    # Calculate discount factors
    discount_factors = [(1 / (1 + discount_rate) ** t) for t in time_periods]
    
    # Initialize DataFrame for RCLV calculations
    df_rclv = pd.DataFrame()
    rclv_values = []
    customer_ids = []
    
    # Calculate RCLV for each customer
    for customer_id in tqdm(active_customers_list):
        customer_ids.append(customer_id)
        s_index = active_customers[active_customers['Customer_ID'] == customer_id]['Renewal_Period_No'].count() - 1
        s_future = survivor_rates_model[s_index:s_index + forecast_horizon]
        
        rclv = sum(
            net_cashflow * discount_factor * survival_future
            for net_cashflow, discount_factor, survival_future in zip(net_cashflows, discount_factors, s_future)
        )
        rclv_values.append(rclv)
        
    # Populate the DataFrame with RCLV results
    df_rclv['Customer_ID'] = customer_ids
    df_rclv['RCLV'] = rclv_values
    df_rclv['RCLV'] = df_rclv['RCLV'].round(1)
    
    return df_rclv
```

```python
# Parameters
survivor_rates_model = mc_sBG_output[1]
discount_rate = 0.05
net_cashflow = 10  # Expected value (or net cashflow) of customer - subscription rate

# Calculate RCLV
rclv_output = calculate_rclv(survivor_rates_model, discount_rate, net_cashflow, selected_cohorts, df, F_horizon)
print('The RCLV table is:')
print(rclv_output)
```

```python
# Merge RCLV and Historical Value outputs on Customer_ID and replace null RCLV values with 0
final_output = pd.merge(hv_output, rclv_output, on='Customer_ID', how='left')
final_output['RCLV'] = final_output['RCLV'].fillna(0)
final_output[['Customer_ID','HV','RCLV','Cohort']]
```
